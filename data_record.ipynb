数据对齐规范注意，选取标准模型之后，M矩阵在这里我们默认置为[1],即单位矩阵。V只与look at相关，注意target的look at位置。如都设置为腰部，那相应的需要记录标准模型的bip位置，在inference和训练的预处理阶段中，需要将bip位置作为look at的输入，而不是look at[0,0,0]，否则数据没有对齐。目前所有面模型都是脚下是[0,0,0]原点。
预处理之后投影到屏幕空间之后，可以视为[v,3]的矩阵，其中v是顶点数量，3是xyz轴分量，特征理解，z是ndc坐标下屏幕空间中的深度值，xyz值任意大于1，都意味着超出屏幕之外，会被cut掉。可以理解为从该视角看下去，所有顶点的分布图和每个像素对应位置的顶点分布图。
